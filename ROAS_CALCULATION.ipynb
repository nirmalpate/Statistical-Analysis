```python
# --- PART 1: A/B TEST DATA ---
# Traffic split 50/50 for Drivalia's CarCloud page
```

```python
visitors_a = 5000 # control group
conv_a = 150 # bookings

visitors_b = 5000 # treatment group
conv_b = 210 # bookings

# Calculations

utilization_rate_a = (conv_a / visitors_a) * 100
utilization_rate_b = (conv_b / visitors_b) * 100

uplift = (utilization_rate_b - utilization_rate_a)/ utilization_rate_a * 100

print(f"utilization_rate_a: {utilization_rate_a}")
print(f"utilization_rate_b: {utilization_rate_b}")
print(f"uplift: {uplift}")
```

```python
price_per_rental  = 200
rental_cost = 100
profit = price_per_rental - rental_cost
profit
gross_margin = (profit / price_per_rental) * 100
gross_margin
profit_on_cost = (profit / rental_cost) * 100
profit_on_cost
print(f"profit {  profit}")
print(f"gross_margin { gross_margin}")
print(f"profit_on_cost { profit_on_cost}")
target_profit_margin = 0.25
target_profit_euro = target_profit_margin * price_per_rental
target_profit_euro
```

```python
Max_Ad_Spend = price_per_rental - rental_cost - target_profit_euro
Max_Ad_Spend

required_roas = (price_per_rental / Max_Ad_Spend) * 100
required_roas
```
